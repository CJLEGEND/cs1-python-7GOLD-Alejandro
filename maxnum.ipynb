```python
# Ask the user to input three numbers
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
num3 = int(input("Enter the third number: "))

# Determine which number is the largest
if num1 >= num2 and num1 >= num3:
  # If num1 is greater than or equal to both num2 and num3
  print ("The largest number is: ",num1)
elif num1 <= num2 and num2 >= num3:
    # If num2 is greater than or equal to both num1 and num3
    print ("The largest number is: ",num2)
else:
 # If neither num1 nor num2 is the largest, num3 must be the largest
  print ("The largest number is: ", num3)
```
